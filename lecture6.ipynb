## JOINT EVENTS AND JOINT PROBABILITY
- Looking at components  or elements of one events that are also present in another event.

```python
## two events
A = {2,3,4}
B = {3,4,6}
SS={1,2,3,4,5,6}

## use a loop

joint_event = [x for x in A if x in B]
joint_event

joint_prob = len(joint_event)/len(SS)
joint_prob
```

## Using the housing dataset
a). Find the probability of getting a house with a basement as well as air conditioning.

```python
import pandas as pd
```

```python
data = pd.read_csv('housing.csv')
data.head()
```

```python
data.columns
```

```python
data_sample_space= data.shape[0]
```

```python

joint_events = data[(data["basement"]=='yes')&(data['airconditioning']=='yes')]
joint_events
```

```python
no_joint_event = joint_events.shape[0]
no_joint_event
```

```python
prob = no_joint_event/data_sample_space
print(f"{prob:.3}")
```
