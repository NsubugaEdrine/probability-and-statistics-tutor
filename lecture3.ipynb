# Measures of dispersion
. range
. std
. var
. min
. max

```python
## min
min_price = housing["price"].min()
min_price
```

```python
## max
max_price = housing["price"].max()
max_price
```

```python
## range
price_range = max_price - min_price
price_range
```

```python
## Standard Deviation
price_std = housing["price"].std()
price_std
```

```python
housing["price"].std() # standard deviation using pandas method
```

```python
## variance
price_variance = housing["price"].var()
price_variance
```

```python
q1 = housing["price"].quantile(0.25) # lower quartile
q2 = housing["price"].quantile(0.75) # upper quartile
iqr = q2 - q1 # interquartile range
iqr
```
