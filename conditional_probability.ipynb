## getting prob of an evnt occuring given a given event has occured(p(A)/p(B))

# calculate the prob of getting a house having AC given has a basement

```python
import pandas as pd
```

```python
data = pd.read_csv('housing.csv')
data
```

```python
data.shape[0]
```

```python

house_with_basement = data[data['basement']=='yes']
house_with_ac_and_basement = house_with_basement[house_with_basement['airconditioning']=='yes']

conditional_prob = house_with_ac_and_basement.shape[0]/house_with_basement.shape[0]
conditional_prob
print(f"{conditional_prob:.3}")
```
