## Mean

## Mode

```python
import numpy as np
data = [23, 40, 35, 23, 40, 40, 40, 21, 25, 38,]

mean = np.mean(data)
print(mean)

median = np.median(data)
print(median)
```

```python
#detailed mode
from scipy import stats
mode_1 = stats.mode(data)
print(mode_1)
```

```python
#mode
import statistics
mode_2 = statistics.mode(data)
print(mode_2)
```

```python
#more about mean and standard deviation
import numpy as np
arr = np.array([10, 20, 30, 40, 50, 60,40])
print("mean: ",np.mean(arr))
print("median: ",np.median(arr))
print("standard deviation: ",np.std(arr))
```
