# breifly about measures of central tendences.
- The mean represents the average value of the dataset.
- The median identifies the central value when the data is ordered.
- The mode shows the most frequently occurring value.



```python
# measures of dispersion
data = [23, 19, 20, 18, 22, 22, 23, 24, 28]
import numpy as np
import pandas as pd
range_value = max(data) - min(data)
variance_value = np.var(data, ddof=1) # sample variance
std_deviation_value = np.std(data, ddof=1) # sample standard deviation  

range_value, variance_value, std_deviation_value
```

```python
# or ds = pd.Series(data)
data = [23, 19, 20, 18, 22, 22, 23, 24, 28]
ds = pd.Series(data)
variance = ds.var()  # variance
std_deviation = ds.std()  # standard deviation      
range_value = ds.max() - ds.min()  # range

variance, std_deviation, range_value
```
