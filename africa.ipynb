## part(a)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('african_development_indicators.csv')

df.head()
```

```python
# i. mean, median, and mode for each of the six numerical variables.
numerical_df = df.select_dtypes(include=['int64', 'float64'])

mean_values = numerical_df.mean()

median_values = numerical_df.median()   

mode_values = numerical_df.mode()
print("Mean values:\n", mean_values)
print("\nMedian values:\n", median_values)
print("\nMode values:\n", mode_values)
```

```python
# ii. mean and median differ noticed in the 'GDP per capita' column, where the mean is significantly higher than the median. This suggests that there are some high-income countries that are skewing the average upwards, while the median provides a better representation of the typical income level for most countries in the dataset.

difference = abs(mean_values - median_values)
print("\nDifference between mean and median for 'GDP per capita':", difference)
```

```python
# iii. Histogram for 'GDP per capita'
gdp = df['GDP_per_Capita_USD']

plt.figure(figsize=(8, 5))
plt.hist(gdp, bins=20)

plt.axvline(gdp.mean(), label ='mean', linestyle ='solid', color='red')
plt.axvline(gdp.median(), label ='median', linestyle ='solid', color='green')

plt.title('Histogram of GDP per Capita')
plt.xlabel('GDP per Capita')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

## part (b)

```python
# i.
range_values = numerical_df.max() - numerical_df.min()
variance_values = numerical_df.var()
std_dev_values = numerical_df.std() 
iqr_values = numerical_df.quantile(0.75) - numerical_df.quantile(0.25)

print("Range values:\n", range_values)  
print("\nVariance values:\n", variance_values)
print("\nStandard Deviation values:\n", std_dev_values)
print("\nInterquartile Range (IQR) values:\n", iqr_values)
```

## part (c)

```python
# i. compute correlation matrix for all six numerical variables.

corr_matrix = numerical_df.corr()
print("Correlation Matrix:\n", corr_matrix) 
```

```python
# display annotated heatmap of the correlation matrix

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# iii. identify from heatmap the pair of variables with the strongest positive correlation 
# and the pair with the strongest negative correlation.
# any pair that shows little or no correlation.

corr_matrix.unstack().sort_values(ascending=False)
```

```python
#iv. scatter plot for the pair of variables with the strongest positive correlation 
# and a scatter plot for the pair with the strongest negative correlation.

#strongest positive correlation
plt.figure(figsize=(6, 4))
sns.scatterplot(x='GDP_per_Capita_USD', y='Life_Expectancy_Years', data=df)
plt.title('GDP vs Life Expectancy')
plt.show()
```
